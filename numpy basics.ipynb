```python
import pandas as pd

# Creating data
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Paris']
}

# Converting dictionary to a Pandas DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# View the first 2 rows of the DataFrame
print(df.head(2))

# Get summary statistics for numeric columns (mean, min, max, etc.)
print(df.describe())

# Check column names and data types
print(df.info())
```

```python
# Select only the 'Age' column
print(df['Age'])

# Filter rows where Age is greater than 28
older_people = df[df['Age'] > 28]
print(older_people)
```

```python
import numpy as np

# Adding a missing value (NaN) for demonstration
df.loc[1, 'Age'] = np.nan 

# Check if there are any missing values
print(df.isnull().sum())

# Fill missing Ages with the average age
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# Calculate the total sales for each Region/Category
sales_data = {
    'Region': ['North', 'North', 'South', 'South'],
    'Product': ['Apple', 'Orange', 'Apple', 'Orange'],
    'Sales': [100, 150, 200, 120]
}
sales_df = pd.DataFrame(sales_data)

# Group by Region and sum the sales
regional_sales = sales_df.groupby('Region')['Sales'].sum()
print(regional_sales)
```
