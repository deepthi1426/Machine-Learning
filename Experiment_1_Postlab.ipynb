```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = sns.load_dataset("titanic")


print(df.head())


print("\nDataset Information:")
print(df.info())


print("\nShape of Dataset:")
print(df.shape)


print("\nMissing Values:")
print(df.isnull().sum())


print("\nDescriptive Statistics:")
print(df.describe())


print("\nMean:")
print(df.mean(numeric_only=True))


print("\nMedian:")
print(df.median(numeric_only=True))


print("\nStandard Deviation:")
print(df.std(numeric_only=True))


print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['age'], bins=20)
plt.title("Age Distribution (24EU01081)")
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=df)
plt.title("Survival Count (24EU01081)")
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='pclass', data=df)
plt.title("Passenger Class Count (24EU01081)")
plt.show()



plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender (24EU01081)")
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Passenger Class (24EU01081)")
plt.show()


plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Heatmap (24EU01081)")
plt.show()
```
